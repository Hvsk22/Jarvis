```python
try:
    WebDriverWait(driver,10).until(EC.presence_of_element_located((By.NAME, "ddlDist")))
finally:
    
    Select(driver.find_element_by_name('ddlDist')).select_by_index(15)
try:
     WebDriverWait(driver,10).until(EC.presence_of_element_located((By.NAME, "ddlAC")))
finally:
     ac=Select(driver.find_element_by_name('ddlAC'))

try:
    WebDriverWait(driver,10).until(EC.presence_of_element_located((By.NAME, "ddlAC")))
finally:
     Select(driver.find_element_by_name('ddlAC')).select_by_index(3)
try:
     WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="btnGetPollingStations"]')))
finally:
     driver.find_element_by_xpath('//*[@id="btnGetPollingStations"]').click()
try:
    WebDriverWait(driver,10).until(EC.presence_of_element_located((By.LINK_TEXT, "View")))
finally:
    links = driver.find_elements_by_link_text('View')
for i in range(len(links)):
    if i%2==0:
        links[i].click()
        time.sleep(5)
        driver.switch_to_window(driver.window_handles[1])
        while len(driver.window_handles) == 2:
            driver.save_screenshot('captcha.png')
            image = Image.open('captcha.png')
            left = 103
            top = 11
            right = 287
            bottom = 60
            image = image.crop((left, top, right, bottom))
            [r,g,b,a] = image.split()
            f = r.convert('RGB')
            f.save('jj.png')
            from PIL import Image
            import pytesseract
            import cv2
            import codecs
            import pandas as pd
            image = cv2.imread('jj.png')
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            cv2.imshow("Image", gray)
            gray = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]

            cv2.imwrite('writen.png', gray)
           
            tessdata_dir_config =  r'--tessdata-dir "C:\Program Files (x86)\Tesseract-OCR\tessdata"'
            pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files (x86)\Tesseract-OCR\tesseract.exe'
            #proc=check_output('RunTess.bat').split()
            text = pytesseract.image_to_string(Image.open('writen.png'),lang = 'eng',config = tessdata_dir_config)

            try:
                    WebDriverWait(driver,10).until(EC.presence_of_element_located((By.NAME, "txtVerificationCode")))
            finally:
                    driver.find_element_by_name('txtVerificationCode').clear()
                    driver.find_element_by_name('txtVerificationCode').send_keys(text)
            try:
                    WebDriverWait(driver,10).until(EC.presence_of_element_located((By.NAME, "btnSubmit")))
            finally:
                    driver.find_element_by_name('btnSubmit').click()
            try:
                 WebDriverWait(driver,10).until(EC.presence_of_element_located((By.ID, "lblCaptchaMessage")))
            finally:
                 cap=driver.find_element_by_id('lblCaptchaMessage')
            if(cap.text=='Please enter correct captcha !'):
                continue
            else:
                driver.close()
            
    else:
        continue

    time.sleep(5)        


```
