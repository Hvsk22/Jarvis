```python
import pandas as pd
import glob
interesting_files = glob.glob('AC-35\*.pdf',recursive= True )

    
```

```python
interesting_files
```

```python
import PyPDF2
```

```python
for i in sorted(interesting_files):
    pdf = open(i,'rb')
    s=i.rfind('\\')
    print(i[s+1:])
    pdfre = PyPDF2.PdfFileReader(pdf)
    x= pdfre.getPage(0)
    pdfWriter = PyPDF2.PdfFileWriter()
    pdfWriter.addPage(x)
    newFile = open(r'C:\Users\User\Desktop\sxxas\krAC-35\kr' + str(i[s+1:]), 'wb')
    pdfWriter.write(newFile)
    newFile.close()
```
