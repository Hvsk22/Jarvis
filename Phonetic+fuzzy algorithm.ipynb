```python
def preprocess(ListOfText):
    temp = []    
    for i in ListOfText:
        name = " ".join(i.split())
        name = re.sub('[.]','',name)
        name =  name.lower()
        temp.append(name)
    return(temp)    
```

```python
def exactmatch(df_list1,df_list):
    out = []
    for i in df_list1:
        c = 0 
        for j in df_list:
            if (i==j):
                out.append("exact")
                c =1
                break
        if(c==0):
            out.append("notfound")
    return(out)        
```

```python
notfound = exactmatch(df_list1,df_list)
```

```python
nf = pd.DataFrame(notfound)
nf.columns = ['match']
#print(len(nf))
#print(df_list1)
booth = pd.DataFrame(df_list1)
booth.columns = ['booth']
matchdf = pd.concat([booth,nf],axis = 1, ignore_index = False)

df_exact = matchdf[(matchdf.match == 'exact')]
df_notfound = matchdf[(matchdf.match == 'notfound')]
#print(len(df_exact))
boothlist = df_notfound.iloc[:,0].tolist()
print((boothlist))
```

```python
def substitute(df):
    output= []
    for i in df:
        c = 0
        words = i.split()
        #print(words)
        final = ""
        for j in words :
            if (j == 'khurd'):
                j = 'kh'
                c=1
            elif(j == 'khur'):
                    j = 'kh'
                    c=1
            elif(j == 'khu'):
                    j = 'kh'
                    c=1
            elif(j == 'bu'):
                j = 'bk'
                c=1
            final = final+j
            
        if (c==0):
            output.append(words[0])
        else:    
            output.append(final.strip())         
    return(output)
    
```

```python
def sound(booth1,village1):
    fo = []
    #fo = pd.DataFrame()
    for i in booth1:
        s=jellyfish.soundex(i)
        #print("Soundex of",i," ",s)
        c = 0
        for j in village1:
            p=jellyfish.soundex(j)
            #print("Soundex of",j," ",p)
            if (s == p):
                #print(type(j))
                #print("inside match")
                fo.append(j)
                c = 1
                break

        if (c == 0):
            fo.append("couldn't find")
    return(fo)       
        
        
```
